```python
X = data.drop("price", axis=1)
y = data['price']
```

```python
X.info()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
numeric_columns = X.select_dtypes(include=['int64', 'float64']).columns
X_numeric = X[numeric_columns]

imputer = SimpleImputer()
X_imputed = imputer.fit_transform(X_numeric)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
pca = PCA(n_components=8)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
X_pca_df = pd.DataFrame(X_pca)
X_pca_df.isna().sum()
```

```python
X_pca_df.head()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
model_lr.fit(X_pca, y)
```

```python
bedrooms = int(input("Enter number of bedrooms: "))
bathrooms = int(input("Enter number of bathrooms: "))
sqft_living = int(input("Enter square footage of living space: "))
sqft_basement = int(input("Enter square footage of basement: "))
sqft_lot = int(input("Enter square footage of lot: "))
sqft_above = int(input("Enter square footage above ground: "))
floors = int(input("Enter number of floors: "))

new_house_data = {
    'bedrooms': [bedrooms],
    'bathrooms': [bathrooms],
    'sqft_living': [sqft_living],
    'sqft_basement': [sqft_basement],
    'sqft_lot': [sqft_lot],
    'sqft_above': [sqft_above],
    'floors': [floors]
}

missing_features = set(X.columns) - set(new_house_data.keys())
for column in missing_features:
    new_house_data[column] = None

new_data_point = pd.DataFrame(new_house_data)

new_data_point = new_data_point[X.columns[X.columns != 'date']]

new_data_point_imputed = imputer.transform(new_data_point)

new_data_point_scaled = scaler.transform(new_data_point_imputed)

new_data_point_pca = pca.transform(new_data_point_scaled)

predicted_price = model_lr.predict(new_data_point_pca)

print("Predicted house price:", predicted_price[0])
```
